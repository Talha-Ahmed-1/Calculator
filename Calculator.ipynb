```python
from tkinter import *
tk=Tk()
tk.title("Calc")
tk.configure(bg="orange")
tk.geometry("265x280")

abcx=""
def press(n):
    global abcx
    if n=="=":
        abc1=eval(abc)
        abcx=str(abc1)
    elif n!="clear":
        abcx=abcx+str(n)
    else:
        abcx=" "
    var.set(abcx)
var=StringVar()


entry=Entry(tk,textvariable=var,justify="right",width=26,bd=20,bg="lightgrey")
entry.grid(row=0,column=0)

clear=Button(tk,text="Clear",fg="black",bg="tan",height=2,width=6,bd=10,command=lambda: press("clear"))
clear.grid(row=0,column=1,sticky=S)
dot=Button(tk,text=".",fg="black",bg="skyblue",height=2,width=6,bd=10,command=lambda: press("."))
dot.grid(row=4,column=0,sticky=W)
equal=Button(tk,text="=",fg="black",bg="pink",height=2,width=6,bd=10,command=lambda: press("="))
equal.grid(row=4,column=0,sticky=E)
button0=Button(tk,text="0",fg="black",bg="skyblue",height=2,width=6,bd=10,command=lambda: press(0))
button0.grid(row=4,column=0)
button1=Button(tk,text="1",fg="black",bg="skyblue",height=2,width=6,bd=10,command=lambda: press(1))
button1.grid(row=3,column=0,sticky=W)
button2=Button(tk,text="2",fg="black",bg="skyblue",height=2,width=6,bd=10,command=lambda: press(2))
button2.grid(row=3,column=0)
button3=Button(tk,text="3",fg="black",bg="skyblue",height=2,width=6,bd=10,command=lambda: press(3))
button3.grid(row=3,column=0,sticky=E)
button4=Button(tk,text="4",fg="black",bg="skyblue",height=2,width=6,bd=10,command=lambda: press(4))
button4.grid(row=2,column=0,sticky=W)
button5=Button(tk,text="5",fg="black",bg="skyblue",height=2,width=6,bd=10,command=lambda: press(5))
button5.grid(row=2,column=0)
button6=Button(tk,text="6",fg="black",bg="skyblue",height=2,width=6,bd=10,command=lambda: press(6))
button6.grid(row=2,column=0,sticky=E)
button7=Button(tk,text="7",fg="black",bg="skyblue",height=2,width=6,bd=10,command=lambda: press(7))
button7.grid(row=1,column=0,sticky=W)
button8=Button(tk,text="8",fg="black",bg="skyblue",height=2,width=6,bd=10,command=lambda: press(8))
button8.grid(row=1,column=0)
button9=Button(tk,text="9",fg="black",bg="skyblue",height=2,width=6,bd=10,command=lambda: press(9))
button9.grid(row=1,column=0,sticky=E)
plus=Button(tk,text="+",fg="black",bg="lightgreen",height=2,width=6,bd=10,command=lambda: press("+"))
plus.grid(row=1,column=1,sticky=E)
minus=Button(tk,text="-",fg="black",bg="lightgreen",height=2,width=6,bd=10,command=lambda: press("-"))
minus.grid(row=2,column=1,sticky=E)
multiply=Button(tk,text="*",fg="black",bg="lightgreen",height=2,width=6,bd=10,command=lambda: press("*"))
multiply.grid(row=3,column=1,sticky=E)
divide=Button(tk,text="/",fg="black",bg="lightgreen",height=2,width=6,bd=10,command=lambda: press("/"))
divide.grid(row=4,column=1,sticky=E)
tk.mainloop()
```
